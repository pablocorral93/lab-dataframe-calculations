## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here

"""We group the dataframe by the column 'CustomerID', calculate the total sum of the amount spent per customerID
   and we store the result in a new dataframa called 'orders_amount_spent'"""

orders_amount_spent = orders.groupby(['CustomerID'], as_index=False).agg({'amount_spent':'sum'})

orders_amount_spent = orders_amount_spent.sort_values(by='amount_spent', ascending=False)

orders_amount_spent
```

```python
#Calculate the 0.95 and 0.75 quantile of the amount spent.

q95 = orders_amount_spent['amount_spent'].quantile(0.95)
q75 = orders_amount_spent['amount_spent'].quantile(0.75)
```

```python
#We define the following function to label the selected costumers.

def label_costumers(row):

  if row['amount_spent'] > q95:
    return 'VIP Customer'

  elif q75 < row['amount_spent'] < q95:

    return 'Preferred customer'
  


orders_amount_spent['Label'] = orders_amount_spent.apply(label_costumers, axis=1)

orders_amount_spent
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here

orders = orders.groupby(['CustomerID','Country'], as_index=False).agg({'amount_spent':'sum'})

orders['Label'] = orders.apply(label_costumers, axis=1)

orders

```

```python
VIP_count = orders[orders['Label'] == 'VIP Customer'].groupby('Country')['CustomerID'].count()

VIP_count

```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here

Preferred_count = orders[orders['Label'] == 'Preferred customer'].groupby('Country')['CustomerID'].count()

Preferred_count
```
